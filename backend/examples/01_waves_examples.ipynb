## 1.1 Square Wave
A square wave is a type of waveform that switches abruptly between two levels — usually a positive and a negative value — with no gradual transition. Unlike a sine wave, which is smooth and continuous, a square wave looks like a series of sharp, rectangular pulses.

They're commonly used in:

- 8-bit video game music
- Digital synthesizers
- Timing circuits and signal testing

```python

SAMPLE_RATE = 44100

def generate_wave(frequency: float, duration: float, volume: float = 0.5) -> np.ndarray:
    if frequency == 0:
        return np.zeros(int(SAMPLE_RATE * duration), dtype=np.float32)
    
    t = np.linspace(0, duration, int(SAMPLE_RATE * duration), endpoint=False)
    wave = volume * np.sign(np.sin(2 * np.pi * frequency * t))
    return wave.astype(np.float32)
```

```python
t = np.linspace(0, 0.5, int(44100 * 0.5), endpoint=False)

plt.figure(figsize=(10, 4))
plt.plot(t[:300], label="Time (t)")
plt.title("Linear Space t")
plt.xlabel("Sample Index")
plt.ylabel("Time (seconds)")
plt.grid(True)
plt.legend()
plt.show()

wave = generate_wave(frequency=440.0, duration=0.5, volume=0.5)

plt.figure(figsize=(10, 4))
plt.plot(wave[:300], label="Square Wave (440Hz)")
plt.title("Square Wave Generated by generate_wave()")
plt.xlabel("Sample Index")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.show()
```
